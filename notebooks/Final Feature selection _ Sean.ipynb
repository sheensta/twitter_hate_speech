```python
import pandas as pd
import numpy as np
from numpy import set_printoptions
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif
```

```python
df = pd.read_csv('tfidf_df.csv')
df.drop(df.columns[0], axis=1,inplace=True)
df
```

```python
X = df.drop(columns = 'class')
Y = df['class']
# feature extraction
test = SelectKBest(score_func=f_classif, k=4)
fit = test.fit(X, Y)
# summarize scores
set_printoptions(precision=3)
print(fit.scores_)
features = fit.transform(X)
# summarize selected features
print(features[0:5,:])
```

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(solver='lbfgs', max_iter = 1000)
rfe = RFE(model, 5)
fit = rfe.fit(X, Y)
print("Num Features: %d" % fit.n_features_)
print("Selected Features: %s" % fit.support_)
print("Feature Ranking: %s" % fit.ranking_)
```

```python
cols = fit.get_support(indices=True)
features_df_new_LR = df.iloc[:,cols]
features_df_new_LR
```

```python
from sklearn.ensemble import RandomForestClassifier


model2 = RandomForestClassifier()
rfe2 = RFE(model2, 100)
fit2 = rfe2.fit(X, Y)
print("Num Features: %d" % fit2.n_features_)
print("Selected Features: %s" % fit2.support_)
print("Feature Ranking: %s" % fit2.ranking_)
```

```python
cols2 = fit2.get_support(indices=True)
features_df_new_RF = df.iloc[:,cols2]
features_df_new_RF
```

```python
rfe3 = RFE(model2, 250)
fit3 = rfe3.fit(X, Y)
print("Num Features: %d" % fit3.n_features_)
print("Selected Features: %s" % fit3.support_)
print("Feature Ranking: %s" % fit3.ranking_)
cols3 = fit3.get_support(indices=True)
features_df_new_RF2 = df.iloc[:,cols3]
features_df_new_RF2
```
